```python
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
def add_ele(*cols):
    i=0
    for elems in cols:
        i+=elems
    return i
df
```

```python
df['sum'] = df.apply(lambda cols: add_ele(cols[0],cols[1],cols[2]),axis=1)
```
