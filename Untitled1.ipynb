```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter([1,2,3,4,5,6],[6,5,4,3,2,1], color='k')
plt.show()
```

```python
slices = [7,2,2,13]
divs = ["education", "food","house rent", "daily needs"]
colors = ["c","m","r","b"]
plt.pie(slices,labels=divs, colors=colors, shadow=False, autopct="%1.1f%%")
plt.legend()
plt.show()
```

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)

t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.02)

plt.subplot(211)
plt.plot(t1, f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))

plt.show()
```
