```python
import pandas as pd
from sklearn.datasets import load_boston 
```

```python
# df = pd.read_csv('train_boston.csv')
boston = load_boston()
print(boston.DESCR)
print(boston.target)
# print(boston.DESCR)
# input_df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
# input_df
```

```python
input_df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
```

```python
input_df
```

```python
input_df.isnull().sum()
```

```python
y = boston.target

y
```

```python
X = pd.get_dummies(input_df, columns=['RAD'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3)
```

```python
print(X_train.shape)

from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, Y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.score(X_train, Y_train)
```

```python
regressor.score(X_test, Y_test)

print(regressor.intercept_ )
```

```python
from sklearn.linear_model import Ridge
regressor2 = Ridge()
regressor2.fit(X_train, Y_train)

regressor2.score(X_train, Y_train)
```

```python
regressor2.score(X_test, Y_test)
```

```python
regressor2.fit(X, y)
```

```python
features = list(X.columns)
coeffs = regressor.coef_

for i, coef in enumerate(coeffs):
    print(features[i], coef)
```

```python
regressor2.score(X,y)
```

```python
y_pred = regressor.predict(X_test)

for idx, y in enumerate(Y_test):
    print('Real: '+ str(y) + ' Predicted: '+ str(y_pred[idx]))
```

```python
new_df = pd.DataFrame(data = {'Real Value': Y_test, 'Predicted Value':y_pred})
new_df.to_csv("result.csv", sep="|", index=False)
```
