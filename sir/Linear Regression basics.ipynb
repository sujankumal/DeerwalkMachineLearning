```python
import numpy as np
```

```python
arr = np.array(((1,1,0), (2, 1, 0), (3, 3, 1), (4,5, 1), (5,4, 1)))
# arr = np.array(((1,2), (2, 3), (3,4), (4,4), (5,5)))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
x = arr[:,:2]
```

```python
x
```

```python
y = arr[:,2]
y
```

```python
y = y.reshape(5, 1)

regressor.fit(x,y)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
regressor.score(x,y)
```
